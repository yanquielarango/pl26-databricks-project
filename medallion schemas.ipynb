##### Create Bronze, Silver, Gold schema




```python
catalog = "dbr_dev"
login = "yanquiel"
storage_account = "dlspl21databricks"

schemas_sql = [
    f"""
    CREATE SCHEMA IF NOT EXISTS `{catalog}`.`{login}_bronze`
    MANAGED LOCATION 'abfss://{login}@{storage_account}.dfs.core.windows.net/bronze'
    """,
    f"""
    CREATE SCHEMA IF NOT EXISTS `{catalog}`.`{login}_silver`
    MANAGED LOCATION 'abfss://{login}@{storage_account}.dfs.core.windows.net/silver'
    """,
    f"""
    CREATE SCHEMA IF NOT EXISTS `{catalog}`.`{login}_gold`
    MANAGED LOCATION 'abfss://{login}@{storage_account}.dfs.core.windows.net/gold'
    """
]

for sql in schemas_sql:
    spark.sql(sql)
```

##### Verify Bronze Schema Location

```python
spark.sql("DESCRIBE SCHEMA EXTENDED dbr_dev.yanquiel_bronze").display()
```
